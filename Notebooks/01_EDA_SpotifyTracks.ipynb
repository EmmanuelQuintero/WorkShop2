### Elimination of the first column
Elimination of the first column
It is eliminated because it is a kind of index and does not allow duplicate data analysis to be performed in the dataframe.

##   Possible transformations

```python
df
```

```python
# Min and Max values of popularity

df['popularity'].min(), df['popularity'].max()
```

```python
# Popularity levels

df['popularity_level'] = df['popularity'].apply(lambda x: 'Baja' if x <= 40 else 'Media' if x < 70 else 'Alta')
```

```python
# Duration in minutes and seconds

df['duration_mins_secs'] = df['duration_ms'].apply(lambda x: f'{int(x/60000)}:{int((x/1000)%60):02d}')
```

```python
# Explicit

df['is_explicit'] = df['explicit'].apply(lambda x: 1 if x == True else 0)
```

```python
df['is_explicit'].value_counts()
```

```python
# is_danceable column (1 if danceability is greater than 0.5, 0 otherwise)

df['is_danceable'] = df['danceability'].apply(lambda x: 0 if x < 0.5 else 1)
```

```python
# is_positive column (1 if valence is greater than 0.5, 0 otherwise)
df['track_feeling'] = df['valence'].apply(lambda x: 'Positive' if x >= 0.5 else 'Negative')
```
