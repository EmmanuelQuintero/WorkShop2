## POSIBLES TRANSFORMACIONES

```python
df
```

```python
# Ver valor minimos y maximos de la columna popularity

df['popularity'].min(), df['popularity'].max()
```

```python
# Crear una nueva columna que sea popularity_level donde si la popularidad es menor a 40 sea baja, si es menor a 70 sea media y si es mayor o igual a 70 sea alta

df['popularity_level'] = df['popularity'].apply(lambda x: 'Baja' if x <= 40 else 'Media' if x < 70 else 'Alta')
```

```python
#Transformar la columna duration_ms a minutos y segundos en una nueva columna llamada duration en formato mm:ss

df['duration_mins_secs'] = df['duration_ms'].apply(lambda x: f'{int(x/60000)}:{int((x/1000)%60):02d}')
```

```python
# Crear una columna is_explicit que sea 1 si la cancion es explicita y 0 si no lo es

df['is_explicit'] = df['explicit'].apply(lambda x: 1 if x == True else 0)
```

```python
df['is_explicit'].value_counts()
```

```python
# Mapear 0 donde danceability este en el rango [0.0 a 0.599] y 1 si esta en el rango [0.600 a 1] en la columna danceability_level

df['is_danceable'] = df['danceability'].apply(lambda x: 0 if x < 0.5 else 1)
```

```python
df['track_feeling'] = df['valence'].apply(lambda x: 'Positive' if x >= 0.5 else 'Negative')
```
